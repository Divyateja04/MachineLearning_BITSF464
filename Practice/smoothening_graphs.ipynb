```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 10, 0.1)  # x axis
z = np.random.normal(0, 50, len(x))  # noise
y = (x ** 2 + z)
window_size = 10

# Apply moving average
y_smooth = np.convolve(y, np.ones(window_size)/window_size, mode='valid')

print(x)
print(x[window_size//2:-window_size//2+1])
print(y)
print(y_smooth)

# Plot smoothed data
plt.plot(x[window_size//2:-window_size//2+1], y_smooth)
plt.show()
```
